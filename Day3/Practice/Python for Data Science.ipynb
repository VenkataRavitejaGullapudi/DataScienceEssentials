### Matplotlib

```
import matplotlib.pyplot as plt
```

```
# Search for more styles
plt.style.use('classic')
```

```
x = np.linspace(0,10,100)
plt.figure()
plt.plot(x,np.sin(x))
# If you put semi column it supress the addres shown above
plt.plot(x,np.cos(x));
```

```
%matplotlib notebook
plt.figure()
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x))
# If you put semi column it supress the addres shown above
plt.plot(x,np.cos(x))
plt.show();
```

```
%matplotlib inline
plt.figure()
x = np.linspace(0,10,100)
plt.plot(x,np.sin(x))
# If you put semi column it supress the addres shown above
plt.plot(x,np.cos(x))
plt.show()
```

```
%matplotlib inline
plt.figure(figsize=(10,5))
x = np.linspace(0,10,100)
plt.subplot(2,1,1)
plt.title("A sin curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.plot(x,np.sin(x))
plt.subplot(2,1,2)

plt.title("A cos curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
# If you put semi column it supress the addres shown above
plt.plot(x,np.cos(x),"--")

plt.show();
```
